```python
import pandas as pd
import sqlite3
```

```python
data_covid = pd.read_csv('covid_19.csv')
```

```python
connection_databse = sqlite3.connect('covid_19.db')
cursor = connection_databse.cursor()
```

```python
cursor.execute("DROP TABLE covid_19")
new_table = """CREATE TABLE covid_19(
                country TEXT,
                continent TEXT,
                population DECIMAL,
                Cases INT,
                Recovered DECIMAL,
                Deaths DECIMAL)"""
```

Sending data to databse

```python
data_covid.to_sql('covid_19',connection_databse,index=False)
```
